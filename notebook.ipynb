```python

import os
import pandas as pd
import numpy as np
import seaborn as sns
import cv2 
from os import listdir
from os.path import isfile, join
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout

%matplotlib inline
```

```python

w = 100
h = 100
l = w * h

def get_image(src):
    img = cv2.imread(src)[:,:,0]
    img = cv2.resize(img, (w,h))
    img = np.array([img])
    img = np.invert(img)
    return img.flatten()

def get_dataset():
    
    columns = np.append(list(range(l)),['digit'])
    df = pd.DataFrame(columns = columns)
    
    folders = [f for f in listdir('sample') ]
    
    for folder in folders:
        curr_folder = 'sample/' + folder
        files = [f for f in listdir(curr_folder) ]
        for file in files:
            image = np.append(get_image(curr_folder +'/'+ file), [int(folder)])
            s = pd.Series(image, index = columns)
            df = df.append(s, ignore_index=True )
    return df
        
df = get_dataset()
```

```python
model = Sequential()
model.add(Dense(l, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(l/2, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(l/4, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(l/16, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['digit'], axis=1)

y = pd.get_dummies(df['digit'], drop_first=False)
y.columns =  ['digit' + '_' + str(x) for x in y.columns]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)

X_train = X_train.values.tolist()
y_train = y_train.values.tolist()
X_test = X_test.values.tolist()
y_test = y_test.values.tolist()

model.fit(x=X_train, y=y_train, epochs=100,batch_size=256, validation_data=(X_test, y_test))
```

```python
from sklearn.metrics import classification_report,confusion_matrix

predictions = model.predict_classes(X_test)
predictions = pd.get_dummies(predictions,drop_first=False)
predictions.columns =  ['digit' + '_' + str(x) for x in predictions.columns]

print(classification_report(y_test,predictions))
```
